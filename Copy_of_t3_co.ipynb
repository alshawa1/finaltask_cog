```python
import pandas as pd
df=pd.read_csv('/content/t33Cogno.csv')
df.head()
```

# **Data Cleaning**:

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
mean_value = df['Item_Weight'].mean()

# Fill null values with the mean
df['Item_Weight'].fillna(mean_value, inplace=True)

print(df)
```

```python
df.isnull().sum()
```

```python
most_frequent_value = df['Outlet_Size'].mode()[0]

# Fill NaN values with the most frequent value
df['Outlet_Size'].fillna(most_frequent_value, inplace=True)

print(df)
```

```python
df.isnull().sum()
```

# **Descriptive Statistics:**

```python
print(df.describe())

# Value counts for categorical columns
print(df['Item_Type'].value_counts())
print(df['Outlet_Size'].value_counts())
print(df['Outlet_Type'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of Item_MRP
plt.figure(figsize=(8, 5))
sns.histplot(df['Item_MRP'], kde=True, bins=30)
plt.title('Distribution of Item MRP')
plt.show()
```

```python
# Bar plot: Item Type Count
plt.figure(figsize=(10, 6))
sns.countplot(y='Item_Type', data=df, order=df['Item_Type'].value_counts().index)
plt.title('Count of Each Item Type')
plt.show()
```

```python
# Box plot: Item_MRP by Outlet_Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outlet_Type', y='Item_MRP', data=df)
plt.title('Item MRP by Outlet Type')
plt.show()
```

```python
# Group by Outlet Size and Outlet Type, and compute average MRP
outlet_analysis = df.groupby(['Outlet_Size', 'Outlet_Type'])['Item_MRP'].mean().reset_index()

print(outlet_analysis)
```

```python
# Assuming you have sales data, group by Outlet_Location_Type
sales_by_location = df.groupby('Outlet_Location_Type')['Item_MRP'].sum().reset_index()

print(sales_by_location)
```

```python
most_sold_items = df['Item_Type'].value_counts().reset_index()
most_sold_items.columns = ['Item_Type', 'Count']
most_sold_items
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y='Item_Type', x='Count', data=most_sold_items.head(10))
plt.title('Top 10 Most Sold Item Types')
plt.show()
```

```python
# Group by Outlet_Type and calculate the mean Item MRP (Maximum Retail Price)
outlet_pricing = df.groupby('Outlet_Type')['Item_MRP'].mean().reset_index()
outlet_pricing.columns = ['Outlet_Type', 'Average_MRP']
outlet_pricing
```

```python
# Group by Outlet_Type and calculate the mean Item MRP (Maximum Retail Price)
outlet_pricing = df.groupby('Outlet_Type')['Item_MRP'].mean().reset_index()

# Now, let's check the columns of outlet_pricing to ensure correctness
print(outlet_pricing.columns)

# Plot the average MRP by outlet type
plt.figure(figsize=(8, 5))
sns.barplot(x='Outlet_Type', y='Item_MRP', data=outlet_pricing)
plt.title('Average Item MRP by Outlet Type')
plt.ylabel('Average MRP')
plt.xlabel('Outlet Type')
plt.show()
```

```python
# Scatter plot to visualize the relationship between Item Visibility and Item MRP
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Item_Visibility', y='Item_MRP', data=df)
plt.title('Item Visibility vs Item MRP')
plt.xlabel('Item Visibility')
plt.ylabel('Item MRP')
plt.show()
```

```python
# Assuming 'Units_Sold' exists, create a total sales column
df['Units_Sold'] = df['Item_Outlet_Sales'] / df['Item_MRP']

# Group by 'Outlet_Identifier' to calculate total sales per outlet
outlet_revenue = df.groupby('Item_Type')['Item_Outlet_Sales'].sum().reset_index()

# Sort to find the outlets with the highest revenue
outlet_revenue = outlet_revenue.sort_values(by='Item_Outlet_Sales', ascending=False)

# Display the top 10 outlets with the highest revenue
top_10_outlets = outlet_revenue.head(10)

print(top_10_outlets)
```

```python
# Group by 'Outlet_Size' and calculate total sales
sales_by_outlet_size = df.groupby('Outlet_Size')['Item_Outlet_Sales'].sum().reset_index()

# Plot sales by outlet size
plt.figure(figsize=(8, 5))
sns.barplot(x='Outlet_Size', y='Item_Outlet_Sales', data=sales_by_outlet_size)
plt.title('Total Sales by Outlet Size')
plt.show()
```

```python
print(df.columns)
```

```python
# Plot distribution of outlet establishment years
plt.figure(figsize=(8, 5))
sns.histplot(df['Outlet_Establishment_Year'], bins=10, kde=True)
plt.title('Distribution of Outlet Establishment Year')
plt.show()
```

```python
# Group by Outlet_Size and Item_Type to count unique item types
outlet_item_variety = df.groupby('Outlet_Size')['Item_Type'].nunique().reset_index()

# Plot item variety by outlet size
plt.figure(figsize=(8, 5))
sns.barplot(x='Outlet_Size', y='Item_Type', data=outlet_item_variety)
plt.title('Variety of Item Types by Outlet Size')
plt.ylabel('Number of Unique Item Types')
plt.show()
```

```python
# Group by Outlet_Location_Type and count unique item types
location_item_variety = df.groupby('Outlet_Location_Type')['Item_Type'].nunique().reset_index()

# Plot item variety by outlet location type
plt.figure(figsize=(8, 5))
sns.barplot(x='Outlet_Location_Type', y='Item_Type', data=location_item_variety)
plt.title('Variety of Item Types by Outlet Location Type')
plt.ylabel('Number of Unique Item Types')
plt.show()
```

```python
# Group by Item_Type to calculate average visibility
item_visibility = df.groupby('Item_Type')['Item_Visibility'].mean().reset_index()

# Plot average visibility by item type
plt.figure(figsize=(10, 6))
sns.barplot(x='Item_Visibility', y='Item_Type', data=item_visibility)
plt.title('Average Visibility by Item Type')
plt.show()
```

```python
# Pivot table to prepare data for heatmap
heatmap_data = df.pivot_table(values='Item_Outlet_Sales', index='Item_Type', columns='Outlet_Size', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.0f')
plt.title('Sales by Item Type and Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Item Type')
plt.show()
```

```python
# Calculate the age of the outlet
df['Outlet_Age'] = 2024 - df['Outlet_Establishment_Year']

# Group by outlet age to calculate average sales
age_sales = df.groupby('Outlet_Age')['Item_Outlet_Sales'].mean().reset_index()

# Plot average sales by outlet age
plt.figure(figsize=(8, 5))
sns.barplot(x='Outlet_Age', y='Item_Outlet_Sales', data=age_sales)
plt.title('Average Sales by Outlet Age')
plt.show()

```

```python
# Group by 'Item_Identifier' to calculate total sales for each item
item_sales = df.groupby('Item_Identifier')['Item_Outlet_Sales'].sum().reset_index()

# Sort to find top contributors
top_items = item_sales.sort_values(by='Item_Outlet_Sales', ascending=False)

# Plot top contributors
plt.figure(figsize=(10, 6))
sns.barplot(x='Item_Outlet_Sales', y='Item_Identifier', data=top_items.head(10))
plt.title('Top 10 Items by Sales Contribution')
plt.xlabel('Total Sales')
plt.ylabel('Item Identifier')
plt.show()
```

```python
# Assuming you have a 'Profit' column
item_profitability = df.groupby('Item_Type')['Units_Sold'].sum().reset_index()

# Sort to find most and least profitable items
item_profitability = item_profitability.sort_values(by='Units_Sold', ascending=False)

# Plot profitability by item type
plt.figure(figsize=(10, 6))
sns.barplot(x='Units_Sold', y='Item_Type', data=item_profitability)
plt.title('Total Profit by Item Type')
plt.xlabel('Total Profit')
plt.ylabel('Item Type')
plt.show()
```

```python
 #Plot price vs sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Item_MRP', y='Item_Outlet_Sales', data=df)
plt.title('Price vs Sales Volume')
plt.xlabel('Item MRP')
plt.ylabel('Item Outlet Sales')
plt.grid(True)
plt.show()
```
